```python
%%time
from scipy.stats import lognorm
from scipy.stats import skewnorm
import math
from scipy.special import erf
  


Positive_01_unx = (Positive_01.data['DF_unexpressed'])
pos_01_unx=[]
for i in Positive_01_unx:
    if i > 0:
        pos_01_unx.append(i)

pos_01_unx_lgm=np.log10(pos_01_unx)
pos_01_unx_mu = np.mean(pos_01_unx_lgm)

pos_01_unx_norm = []
for i in pos_01_unx_lgm:
    if i > 0:
        c = i/pos_01_unx_mu
        pos_01_unx_norm.append(c)
        
Positive_02_unx = (Positive_02.data['DF_unexpressed'])
pos_02_unx=[]
for i in Positive_02_unx:
    if i > 0:
        pos_02_unx.append(i)
        
pos_02_unx_lgm=np.log10(pos_02_unx)
pos_02_unx_mu = np.mean(pos_02_unx_lgm)

pos_02_unx_norm = []
for i in pos_02_unx_lgm:
    if i > 0:
        c = i/pos_02_unx_mu
        pos_02_unx_norm.append(c)
        
Positive_03_unx = (Positive_03.data['DF_unexpressed'])
pos_03_unx=[]
for i in Positive_03_unx:
    if i > 0:
        pos_03_unx.append(i)
        
pos_03_unx_lgm=np.log10(pos_03_unx)
pos_03_unx_mu = np.mean(pos_03_unx_lgm)

pos_03_unx_norm = []
for i in pos_03_unx_lgm:
    if i > 0:
        c = i/pos_03_unx_mu
        pos_03_unx_norm.append(c)
        

Positive_04_unx = (Positive_04.data['DF_unexpressed'])
pos_04_unx=[]
for i in Positive_04_unx:
    if i > 0:
        pos_04_unx.append(i)
        
pos_04_unx_lgm=np.log10(pos_04_unx)
pos_04_unx_mu = np.mean(pos_04_unx_lgm)

pos_04_unx_norm = []
for i in pos_04_unx_lgm:
    if i > 0:
        c = i/pos_04_unx_mu
        pos_04_unx_norm.append(c)
        
pos_unx_total=pos_01_unx_norm+pos_02_unx_norm+pos_03_unx_norm+pos_04_unx_norm
      

Positive_01_ex = (Positive_01.data['DF_expressed'])
pos_01_ex=[]
for i in Positive_01_ex:
    if i > 0:
        pos_01_ex.append(i)

pos_01_ex_lgm=np.log10(pos_01_ex) 

pos_01_ex_norm = []
for i in pos_01_ex_lgm:
    if i > 0:
        c = i/pos_01_unx_mu
        pos_01_ex_norm.append(c)
        
Positive_02_ex = (Positive_02.data['DF_expressed'])
pos_02_ex=[]
for i in Positive_02_ex:
    if i > 0:
        pos_02_ex.append(i)

pos_02_ex_lgm=np.log10(pos_02_ex)

pos_02_ex_norm = []
for i in pos_02_ex_lgm:
    if i > 0:
        c = i/pos_02_unx_mu
        pos_02_ex_norm.append(c)
        
Positive_03_ex = (Positive_03.data['DF_expressed'])
pos_03_ex=[]
for i in Positive_03_ex:
    if i > 0:
        pos_03_ex.append(i)

pos_03_ex_lgm=np.log10(pos_03_ex)


pos_03_ex_norm = []
for i in pos_03_ex_lgm:
    if i > 0:
        c = i/pos_03_unx_mu
        pos_03_ex_norm.append(c)
        

Positive_04_ex = (Positive_04.data['DF_expressed'])
pos_04_ex=[]
for i in Positive_04_ex:
    if i > 0:
        pos_04_ex.append(i)

pos_04_ex_lgm=np.log10(pos_04_ex)


pos_04_ex_norm = []
for i in pos_04_ex_lgm:
    if i > 0:
        c = i/pos_04_unx_mu
        pos_04_ex_norm.append(c)
        
pos_ex_total=pos_01_ex_norm+pos_02_ex_norm+pos_03_ex_norm+pos_04_ex_norm  


sigma = np.std(pos_ex_total)
mu = np.mean(pos_ex_total)
print("mu =",mu)

sigma2 = (np.std(pos_unx_total))
mu2 = np.mean(pos_unx_total)
print("mu2 =",mu2)
            

num_bins = np.linspace(0,1.5,500)


fig, ax = plt.subplots(figsize=(15,10))

# histogram of the data
n, bins, patches = ax.hist((pos_ex_total), num_bins, density=True)

n2, bins2, patches2 = ax.hist((pos_unx_total), num_bins, density=True, alpha = 0.5)


# add Gaussians 
p = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
ax.plot(bins, p, '--')

p2 = ((1 / (np.sqrt(2 * np.pi) * sigma2)) *
    np.exp(-0.5 * (1 / sigma2 * (bins2 - mu2))**2))
ax.plot(bins2, p2,'--')



#Calc Area of Gaussian Using Trapezoidal Apdaptive Quadrature & Solve for 99.5% at some to-be determined x value
def traparea(f,a,b):
    h = (b-a)
    area = h*(f(a)+f(b))/2
    return area

def adapt(f,a,b,tol=0.00005 ):
    m = (b+a)/2
    area = 0
    areatot = traparea(f,a,b)
    nextareatot = traparea(f,a,m) + traparea(f,m,b)
    err = np.abs(areatot-nextareatot)
    
    if err < tol:
        return areatot
    
    else:
        arealeft = adapt(f,a,m,tol/2)
        arearight = adapt(f,m,b,tol/2)
        area = area + arealeft + arearight
        
    return area


def function(x):
    return (1 / (np.sqrt(2 * np.pi) * sigma2)) * np.exp(-0.5 * (1 / sigma2 * (x - mu2))**2)  


# Solve for x value

for i in np.arange(1, 1.50,.0005):
    f = function
    current_range_a = 0
    current_range_b = i
    adapt(f,current_range_a,current_range_b)
    
    if adapt(f,current_range_a,current_range_b) >=.990:
        break
        
        
print('% Area at x =',current_range_b,'is',adapt(f,current_range_a,current_range_b)*100,'%')


#Add line at 99.0 % Gaussian

x1, y1 = [current_range_b,current_range_b], [0,30]
ax.plot(x1,y1,color="red",linewidth=2, alpha = 0.75, label = "Gaussian 99.0%")


ax.set_title('Histogram of DF Intensity', fontsize = 25)
ax.set_xlim(0.9,1.3)
#ax.set_ylim(3.5,5)
ax.legend(fontsize=25)
ax.set_xlabel("Log Intensity",fontsize=25)
ax.set_ylabel("Density",fontsize=25)
plt.minorticks_on()

# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()
plt.show()
fig.savefig(o+'pos.jpg')
```

```python
%%time
from scipy.stats import lognorm
from scipy.stats import skewnorm
import math
from scipy.special import erf
   

sigma = np.std(ctrl_ex_total)
mu = np.mean(ctrl_ex_total)
print("mu =",mu)

sigma2 = (np.std(ctrl_unx_total))
mu2 = np.mean(ctrl_unx_total)
print("mu2 =",mu2)
            

num_bins = np.linspace(0,1.5,1000)


fig, ax = plt.subplots(figsize=(15,10))

# histogram of the data
n, bins, patches = ax.hist((ctrl_ex_total), num_bins, density=True)

n2, bins2, patches2 = ax.hist((ctrl_unx_total), num_bins, density=True, alpha = 0.5)


# add Gaussians 
p = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
ax.plot(bins, p, '--')

p2 = ((1 / (np.sqrt(2 * np.pi) * sigma2)) *
    np.exp(-0.5 * (1 / sigma2 * (bins2 - mu2))**2))
ax.plot(bins2, p2,'--')



#Calc Area of Gaussian Using Trapezoidal Apdaptive Quadrature & Solve for 99.5% at some to-be determined x value
def traparea(f,a,b):
    h = (b-a)
    area = h*(f(a)+f(b))/2
    return area

def adapt(f,a,b,tol=0.00005 ):
    m = (b+a)/2
    area = 0
    areatot = traparea(f,a,b)
    nextareatot = traparea(f,a,m) + traparea(f,m,b)
    err = np.abs(areatot-nextareatot)
    
    if err < tol:
        return areatot
    
    else:
        arealeft = adapt(f,a,m,tol/2)
        arearight = adapt(f,m,b,tol/2)
        area = area + arealeft + arearight
        
    return area


def function(x):
    return (1 / (np.sqrt(2 * np.pi) * sigma2)) * np.exp(-0.5 * (1 / sigma2 * (x - mu2))**2)  


# Solve for x value

for i in np.arange(1, 1.50,.0005):
    f = function
    current_range_a = 0
    current_range_b = i
    adapt(f,current_range_a,current_range_b)
    
    if adapt(f,current_range_a,current_range_b) >=.990:
        break
        
        
print('% Area at x =',current_range_b,'is',adapt(f,current_range_a,current_range_b)*100,'%')


#Add line at 99.0 % Gaussian

x1, y1 = [current_range_b,current_range_b], [0,30]
ax.plot(x1,y1,color="red",linewidth=2, alpha = 0.75, label = "Gaussian 99.0%")


ax.set_title('Histogram of DF Intensity', fontsize = 25)
ax.set_xlim(0.9,1.3)
#ax.set_ylim(3.5,5)
ax.legend(fontsize=25)
ax.set_xlabel("Log Intensity",fontsize=25)
ax.set_ylabel("Density",fontsize=25)
plt.minorticks_on()

# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()
plt.show()
fig.savefig(o+'neg.jpg')
```
